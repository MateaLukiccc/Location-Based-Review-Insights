## Sredjivanje podataka

```python
import re

def clean_english_text(text):
    """
    Extracts English text and cleans formatting from a mixed-language string.
    Handles NaN values gracefully.
    
    Args:
        text (str): Input text that may contain translated and original sections
        
    Returns:
        str: Cleaned English text, or empty string for NaN values
    """
    if pd.isna(text):
        return ""
    
    text = str(text)
    if "(Translated by Google)" in text:
        match = re.search(r'\(Translated by Google\)(.*?)(?=\(Original\)|$)', text, re.DOTALL)
        if match:
            english_text = match.group(1)
        else:
            english_text = text
    else:
        english_text = text
    
    cleaned_text = (english_text
                   .strip()                  
                   .replace('\n', ' ')        
                   .replace('\r', ' ')        
                   .replace('\t', ' ')         
                   )
    
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)
    
    return cleaned_text

df_kotor["review_text"] = df_kotor["review_text"].apply(clean_english_text)
df_kotor["review_text"].head(15)
```

## Vizualizacija

```python
df_kotor_copy = df_kotor.copy()

df_kotor_copy['Text_Length'] = df_kotor_copy['review_text'].str.len()
df_kotor_copy['Word_Count'] = df_kotor_copy['review_text'].str.split().str.len()
df_kotor_copy.head()
```

```python
df_kotor_copy['review_text'] = df_kotor_copy['review_text'].str.lower()

sns.displot(df_kotor_copy['Text_Length'])
plt.xlim([0, 512])
```

```python
sns.boxplot(x="review_rating", y="Text_Length", data=df_kotor_copy)
```

```python
sns.boxplot(x="review_rating", y="Word_Count", data=df_kotor_copy)
```

```python
count = df_kotor_copy['review_rating'].value_counts()
background_color = '#ffffff'
sns.set_theme(style="whitegrid", rc={"axes.facecolor": background_color, 'figure.facecolor': background_color})


fig, axs = plt.subplots(1, 2, figsize=(12, 6), facecolor=background_color)
palette = sns.color_palette("bright", len(count))
sns.set_palette(palette)
axs[0].pie(count, labels=count.index, autopct='%1.1f%%', startangle=140)
axs[0].set_title('Raspodela po ocenama', fontsize=15, fontweight='bold')

sns.barplot(x=count.index, y=count.values, ax=axs[1], palette=palette)
axs[1].set_title('Zastupljenost ocena', fontsize=15, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud

df_5_star = df_kotor_copy[df_kotor_copy['review_rating']==5]
df_4_star = df_kotor_copy[df_kotor_copy['review_rating']==4]
df_3_star = df_kotor_copy[df_kotor_copy['review_rating']==3]
df_2_star = df_kotor_copy[df_kotor_copy['review_rating']==2]
df_1_star = df_kotor_copy[df_kotor_copy['review_rating']==1]

combined_5_text = ' '.join(df_5_star['review_text'])
combined_4_text = ' '.join(df_4_star['review_text'])
combined_3_text = ' '.join(df_3_star['review_text'])
combined_2_text = ' '.join(df_2_star['review_text'])
combined_1_text = ' '.join(df_1_star['review_text'])

s5_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_5_text)
s4_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_4_text)
s3_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_3_text)
s2_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_2_text)
s1_wordcloud = WordCloud(width=800, height=400, background_color='white').generate(combined_1_text)

plt.figure(figsize=(18, 9))

plt.subplot(2, 3, 1)
plt.imshow(s5_wordcloud, interpolation='bilinear')
plt.title('5 star')
plt.axis('off')

plt.subplot(2, 3, 2)
plt.imshow(s4_wordcloud, interpolation='bilinear')
plt.title('4 star')
plt.axis('off')

plt.subplot(2, 3, 3)
plt.imshow(s3_wordcloud, interpolation='bilinear')
plt.title('3 star')
plt.axis('off')

plt.subplot(2, 3, 4)
plt.imshow(s2_wordcloud, interpolation='bilinear')
plt.title('2 star')
plt.axis('off')

plt.subplot(2, 3, 5)
plt.imshow(s1_wordcloud, interpolation='bilinear')
plt.title('1 star')
plt.axis('off')

plt.tight_layout()
plt.show()
```
